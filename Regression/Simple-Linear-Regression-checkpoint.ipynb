# Simple Linear Regression

```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset

dataset = pd.read_csv('/Users/palashchaudhari/Desktop/Salary_Data.csv')
print(dataset)
```

```python
#shaping and assigning

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
#split dataset into trainig and test data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
#train the simple regression model on train dataset

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
#visualization on training dataset

plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary Vs Exp - Training Dataset')
plt.xlabel('Years of Exp')
plt.ylabel('Salary')
plt.show()
```

```python
#visualization on test dataset

plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary Vs Exp - Training Dataset')
plt.xlabel('Years of Exp')
plt.ylabel('Salary')
plt.show()
```
